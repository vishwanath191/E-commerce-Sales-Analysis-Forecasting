```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from prophet import Prophet


pd.set_option("display.max_columns",None)

from google.colab import files
uploaded = files.upload()


df = pd.read_csv("data.csv", encoding="ISO-8859-1")
df.head()

df.info()
df.isnull().sum()


df = df.dropna(subset=["CustomerID"])


df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])


df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

df.head()




df["MonthYear"] = df["InvoiceDate"].dt.to_period("M")
monthly_revenue = df.groupby("MonthYear")["TotalPrice"].sum()

plt.figure(figsize=(12,6))
monthly_revenue.plot(kind="bar")
plt.title("Monthly Revenue")
plt.ylabel("Revenue")
plt.show()


country_sales = df.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False)[:10]
sns.barplot(x=country_sales.values, y=country_sales.index)
plt.title("Top 10 Countries by Sales")
plt.show()


# RFM Analysis
# Calculate Recency, Frequency, and Monetary
latest_date = df["InvoiceDate"].max()
rfm = df.groupby("CustomerID").agg(
    Recency=("InvoiceDate", lambda date: (latest_date - date.max()).days),
    Frequency=("InvoiceNo", "nunique"),
    Monetary=("TotalPrice", "sum")
)

# Scale RFM values
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

kmeans = KMeans(n_clusters=4, random_state=42)
rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)

sns.scatterplot(data=rfm, x="Recency", y="Monetary", hue="Cluster", palette="viridis")
plt.title("Customer Segments")
plt.show()
# 7️⃣ Forecasting with Prophet

# Aggregate daily sales
daily_sales = df.groupby("InvoiceDate")["TotalPrice"].sum().reset_index()
daily_sales.rename(columns={"InvoiceDate": "ds", "TotalPrice": "y"}, inplace=True)

# Fit model
model = Prophet()
model.fit(daily_sales)

# Future dataframe
future = model.make_future_dataframe(periods=90)  # Forecast 90 days
forecast = model.predict(future)

# Plot forecast
fig = model.plot(forecast)
plt.title("Sales Forecast (Next 3 Months)")
plt.show()
# 8️⃣ Save Results
# Save RFM clusters
rfm.to_csv("rfm_segments.csv")

# Save forecast results
forecast.to_csv("sales_forecast.csv")

from google.colab import files
files.download("rfm_segments.csv")
files.download("sales_forecast.csv")
```
